```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
venues = pd.read_csv("D:/EM_38/data/venues_clean.csv")
vendors = pd.read_csv("D:/EM_38/data/vendors_clean.csv")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(venues["avg_price"], bins=30, kde=True)

plt.title("Venue Average Price Distribution")
plt.xlabel("Average Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(vendors["avg_price"], bins=30, kde=True)

plt.title("Vendor Average Price Distribution")
plt.xlabel("Average Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

vendors["listing_type"].value_counts().plot(kind="bar")

plt.title("Vendor Category Distribution")
plt.xlabel("Vendor Type")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(vendors["rating"], bins=20, kde=True)

plt.title("Vendor Rating Distribution")
plt.xlabel("Rating")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=vendors,
    x="avg_price",
    y="rating",
    hue="listing_type"
)

plt.title("Average Price vs Rating")

plt.show()
```

```python
plt.figure(figsize=(8,5))

venues["city"].value_counts().head(10).plot(kind="bar")

plt.title("Top Cities by Venue Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

vendors["city"].value_counts().head(10).plot(kind="bar")

plt.title("Top Cities by Vendor Count")

plt.show()
```

```python
vendors.groupby("listing_type")["avg_price"].mean().sort_values()
```

```python
vendors.groupby("listing_type")["rating"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))

vendors.groupby("listing_type")["avg_price"].mean().sort_values().plot(kind="bar")

plt.title("Average Price by Vendor Type")
plt.ylabel("Average Price")

plt.show()
```

```python
plt.figure(figsize=(8,5))

vendors.groupby("listing_type")["rating"].mean().sort_values().plot(kind="bar")

plt.title("Average Rating by Vendor Type")

plt.show()
```
